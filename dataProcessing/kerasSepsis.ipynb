```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
dataset = pd.read_csv('sepDataSet2.csv', header=None)

dataset.head(2)
```

```python
dataset.describe(include='all')
```

```python
sns.heatmap(dataset.corr(), annot=True)
```

```python
# creating input features and target variables
X= dataset.iloc[:,0:120]
y= dataset.iloc[:,120]
```

```python
X.head(12)
```

```python
#standardizing the input feature
# from sklearn.preprocessing import StandardScaler
# sc = StandardScaler()
# X = sc.fit_transform(X)
# X
```

```python
from sklearn.model_selection import train_test_split
#shuffled
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3)
```

```python
from keras import Sequential
from keras.layers import Dense
```

```python
classifier = Sequential()
#First Hidden Layer
classifier.add(Dense(60, activation='relu', kernel_initializer='random_normal', input_dim=120))
#Second  Hidden Layer
classifier.add(Dense(60, activation='relu', kernel_initializer='random_normal'))
#Output Layer
classifier.add(Dense(1, activation='sigmoid', kernel_initializer='random_normal'))
```

```python
#Compiling the neural network
classifier.compile(optimizer ='adam',loss='binary_crossentropy', metrics =['accuracy'])
```

```python
#Fitting the data to the training dataset
classifier.fit(X_train,y_train)
#optional parameters
#batch_size=10, epochs=100

#KERAS CLASSIFICATION ACCURACY REPORTED 75%
```

```python
#loss value & metrics values
eval_model=classifier.evaluate(X_train, y_train)
eval_model
```

```python
#prediction is greater than 0.5 then the output is 1 else the output is 0
y_pred=classifier.predict(X_test)
y_pred =(y_pred>0.5)
```

```python
#going across
#true pos, false pos, false neg, true neg
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
print(cm)
```
